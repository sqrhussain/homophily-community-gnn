```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys
sys.path.insert(0,'..')
from src.misc.separator import splitDataFrameList
from scipy.stats import ttest_rel
```

```python
def load_model_results(models,datasets):
    dfs = []
    for model in models:
        for dataset in datasets:
            df = pd.read_csv(f'../reports/results/test_acc/{model}_{dataset}.csv')
            df['dataset'] = dataset
            dfs.append(df[(df.splits == 100) & (df.inits == 20)])
    return pd.concat(dfs,ignore_index=True)
```

```python
datasets = 'texas wisconsin cornell washington'.split()
df_results = load_model_results('gcn'.split(),datasets)
view_columns = 'dataset conv arch test_accs'.split()
df_results = df_results[view_columns].rename(columns={'test_accs':'test_acc','arch':'directionality'})
df_results
```

```python
for i in range(4):
    A = np.array([float(x) for x in df_results.loc[i*3+0,'test_acc'][1:-1].split(', ')])
    B = np.array([float(x) for x in df_results.loc[i*3+1,'test_acc'][1:-1].split(', ')])
    C = np.array([float(x) for x in df_results.loc[i*3+2,'test_acc'][1:-1].split(', ')])
    print(f'{ttest_rel(A,B)},\n{ttest_rel(A,C)},\n{ttest_rel(B,C)}\n\n')
```
